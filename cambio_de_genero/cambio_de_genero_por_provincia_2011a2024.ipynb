```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cambio_genero_provincia_genero_edad_mayo_2024.csv")
```

```python
df.head(15)
```

```python
df.info()
```

```python
df.drop(["provincia_id"],axis=1, inplace = True)
```

```python
df.head()
```

```python
df = df.rename(columns = {"provincia_nombre":"provincia", "edad_actual_grandes_grupos":"grupo_edad"})
```

```python
df['grupo_edad'] = pd.Categorical(
    df['grupo_edad'],
    categories=['0 a 17','18 a 29','30 a 39','40 y mas'],
    ordered=True
)
```

Categorización de los datos para un mejor análisis

```python
df["provincia"] = df["provincia"].str.strip().astype('category')
```

```python
total_personas=df["cantidad"].sum()
print(total_personas)
```

Total de personas que realizaron cambio de género desde 2011 a 2024

```python
df.groupby("provincia")["cantidad"].sum().sort_values(ascending=False)
```

Total de personas que hicieron cambio de género por cada provincia

```python
df.groupby("genero")["cantidad"].sum()
```

Total de los 3 géneros de trámite. La mayoría de los trámites son cambio a género F, seguido por M y último a X.

```python
prov_selec = ["Buenos Aires", "Ciudad Autónoma de Buenos Aires", "Santa Fe", "Córdoba", "Salta"]

df_filt = df[df["provincia"].isin(prov_selec)]

df_filt["provincia"] = df_filt["provincia"].cat.remove_unused_categories()

df_filt.groupby ("provincia")["cantidad"].sum().sort_values(ascending=False).plot(kind="bar")

plt.title("Total personas")
plt.xlabel("")
plt.xticks(rotation=80)
plt.ylabel("Total")
plt.show()
```

Las 5 principales Provincias en cuanto a cantidad de trámites.

```python
df.groupby("grupo_edad")["cantidad"].sum().sort_values(ascending=False)
```

Total de personas que realizaron el trámite de cambio de género según grupo de edad

```python
df.groupby("grupo_edad")["cantidad"].sum().sort_values(ascending=False).plot(kind="bar")

plt.title("Total personas por grupo edad")
plt.xlabel("")
plt.xticks(rotation=45)
plt.ylabel("Total")
plt.show()
```

Representación del total en todo el país según la edad 

```python
df_ba=df[df["provincia"] == "Buenos Aires"]

df_ba.groupby("grupo_edad")["cantidad"].sum().sort_values(ascending=False).plot(kind="bar")

plt.title("Total Bs As por grupo edad")
plt.xlabel("Grupo de edad")
plt.xticks(rotation=45)
plt.ylabel("Total")
plt.show()
```

En la Provincia de Buenos Aires.

```python
prov_top = ["Buenos Aires", "Ciudad Autónoma de Buenos Aires", "Santa Fe", "Córdoba", "Salta"]

df_top = df[df["provincia"].isin(prov_top)]

df_top["provincia"] = df_top["provincia"].cat.remove_unused_categories()

df_top.pivot_table(index="grupo_edad", columns="provincia", values="cantidad", aggfunc="sum").plot(kind="bar", stacked=True)


plt.title("Distribución por Provincia")
plt.xlabel("Grupo de edad")
plt.xticks(rotation=45)
plt.ylabel("Total Personas")
plt.legend(title="Provincia",loc ="upper right", fontsize=8, frameon=False)
plt.show()
```

Clasificación de cambios de género según provincia (las 5 principales en cuanto a cantidad de trámites) y grupo de edad en base al total de las personas que lo realizaron.

```python
prov_top = ["Buenos Aires", "Ciudad Autónoma de Buenos Aires", "Santa Fe", "Córdoba", "Salta"]
df_top = df[df["provincia"].isin(prov_top)].copy()

fig, axes = plt.subplots(1, 5, figsize=(20,5))

for ax, prov in zip(axes, prov_top):
    datos = df_top[df_top["provincia"] == prov].groupby("grupo_edad")["cantidad"].sum()
    ax.pie(datos, labels=datos.index, autopct="%1.1f%%")
    ax.set_title(prov)

plt.suptitle("Distribución por grupo de edad en cada provincia (top 5)")
plt.show()
```

En éste gráfico se clasifican en porcentajes los cambios de género segun la edad en cada una de las primeras 5 provincias.

Concluimos el análisis exploratorio y visualizaciones de los datos aportados por el Gobierno Nacional en cuanto a los trámites de cambio de género en el DNI de personas entre los años 2011 a 2024.
